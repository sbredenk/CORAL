# Turbines

```python
years = np.arange(2022,2041)
width=0.3
fig, ax = plt.subplots(1,1, sharex=True, figsize=(12,6))
ax.bar(years-width, tier1_low['Nacelles'], width)
ax.bar(years, tier1_mid['Nacelles'], width)
ax.bar(years+width, tier1_high['Nacelles'], width)
ax.set_ylabel('Turbines')
# axes[0].fill_between(tier1_high.index, tier1_high['Blades'], tier1_low['Blades'], alpha=0.2)
# axes[1].plot(tier1_mid['Nacelles'])
# axes[1].plot(tier1_low['Nacelles'])
# axes[1].plot(tier1_high['Nacelles'])
# axes[1].set_ylabel('Nacelles')
# axes[2].plot(tier1_mid['Towers'])
# axes[2].plot(tier1_low['Towers'])
# axes[2].plot(tier1_high['Towers'])
# axes[2].set_ylabel('Towers')
ax.legend(['Base','Low','High'])
ax.set_xticks(years)
plt.show()
fig.savefig('turbines', bbox_inches='tight')
```

# Foundations

```python
years = np.arange(2022,2041)
width=0.3
fig, axes = plt.subplots(4,1, figsize=(12,10), sharex=True)
axes[0].bar(years - width, tier1_low['Monopile Foundations'], width)
axes[0].bar(years, tier1_mid['Monopile Foundations'], width)
axes[0].bar(years + width, tier1_high['Monopile Foundations'], width)
axes[0].set_ylabel('Monopiles')
# axes[0].set_xticks(years)
# axes[0].fill_between(tier1_high.index, tier1_high['Blades'], tier1_low['Blades'], alpha=0.2)
axes[1].bar(years - width, tier1_low['Jacket Foundations'], width)
axes[1].bar(years, tier1_mid['Jacket Foundations'], width)
axes[1].bar(years + width, tier1_high['Jacket Foundations'], width)
axes[1].set_ylabel('Suction-bucket Jackets')
# axes[1].set_xticks(years)
axes[2].bar(years - width, tier1_low['Gravity-based Foundations'], width)
axes[2].bar(years, tier1_mid['Gravity-based Foundations'], width)
axes[2].bar(years + width, tier1_high['Gravity-based Foundations'], width)
axes[2].set_ylabel('GBFs')
# axes[2].set_xticks(years)
axes[3].bar(years - width, tier1_mid['Semisubmersible Platform'], width)
axes[3].bar(years, tier1_low['Semisubmersible Platform'], width)
axes[3].bar(years + width, tier1_high['Semisubmersible Platform'], width)
axes[3].set_ylabel('Semisubs')
axes[0].legend(['Low','Base','High'])

axes[3].set_xticks(years)
plt.show()
fig.savefig('foundations', bbox_inches='tight')
```
